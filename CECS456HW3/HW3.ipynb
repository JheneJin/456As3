```python
import numpy as np
import matplotlib.pyplot as plt

#return the math of the sigmoid function
def sigmoid(t):
    return 1/(1+ np.exp(-t))

#take in parameters x and y values, k being the iteration
def gradDescent(x, y, k):
    #set the values of w0, w1, and alpha
    w0 = 3
    w1 = 3
    alpha = 1
    #set an xRange so that it creates the sample data for a 100 points
    xRange = np.linspace(min(x), max(x), 100)

    
    #creates empty arrays for w0, w1 and yRange
    w0Arr = []
    w1Arr = []
    yRangeArr = []

    #iterate through k times
    for training in range(k):
        #reset gradient w0 and w1 every timme it gets back here
        gradW0 = 0
        gradW1 = 0
        #iterate though the length of x
        for i in range(len(x)):
            #extract the individual xI and yI from the x and y data values
            xI = x[i]
            yI = y[i]
            #predicts the y value depending on weights
            yHatI = sigmoid(w1 * xI + w0)
            #calculate the gradW1 and gradW0
            gradW1 = -(yI -yHatI) * xI
            gradW0 = -(yI -yHatI)
            #caluclate the w0 and w1
            w0 -= alpha * gradW0
            w1 -= alpha * gradW1
            #append it into the arrays
            w0Arr.append(w0)
            w1Arr.append(w1)
        #calculate the Y-Range
        yRange = sigmoid(w1* xRange + w0)
        #appends the Y-Range
        yRangeArr.append(yRange)
    #call the graph function
    graph(x, y, xRange, yRange)

#takes the inputs x, y, xRange, yRange
def graph(x, y, xRange, yRange):
    #plot the scatter plots fromm the 
    plt.scatter(x, y, label='Data Points')
    #plot the cross entropy
    plt.plot(xRange, yRange, color='red', label='Logistic Regression Line')
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.legend()
    plt.show()

#filePath
dataPath = '2d_classification_data_v1_entropy.csv'
#load the file
data = np.loadtxt(dataPath, delimiter = ',')
#extract x and y
x = data[:-1,:]
y = data[-1,:]
# get the initial x values from the double martrix
xData = x[0] 

gradDescent(xData, y, 25)

```
